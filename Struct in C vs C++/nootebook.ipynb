#struct in C

```python
struct Student {
    char name[50];
    int age;
};

// to access struct's elements: functions must be external and accept the struct

void printStudent(struct Student s) {
printf("%s is %d\n", s.name, s.age);
}

// must use struct keyword  in main
int main() {
    struct Student s1 = {"Alice", 20};
    printStudent(s1);
}
```

#struct in C++

```python
struct Student {
    char name[50];
    int age;
    //struct can have member functions
    void print() {
        cout << name << " is " << age << endl;
        }
};

// in c++ no struct keyword needed for main, struct name is a type
int main() {
    Student s1(“Alice”
    , 20);
    s1.print(); // Object-oriented call
    }
}
```
